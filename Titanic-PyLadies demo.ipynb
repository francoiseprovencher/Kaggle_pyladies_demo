## Preparing the data for a machine learning classifier

The classifier we will use takes an array as an input, so we need to do three things:

- Fill in the **missing values**
- Transform non-numeric data into **numeric data** (e.g. male/female to 0/1)
- **Drop** the columns of data we will not use

### 1 - Fill in the missing values

I think age is an important feature, so I will fill in the missing age values with the median age of people on board. You can make fancier assumptions based on gender and class, but this is just a simple example.

```python
#To select a column (e.g. 'Age') for missing values, use
df[df['Age'].isnull()][['Sex', 'Pclass', 'Age']]
```

```python
#Get the median age of the passengers
median_age = df['Age'].median()
print median_age
```

```python
#Make a new column named 'AgeFill'
df['AgeFill'] = df['Age']
df.head(8)
```

```python
# Now we fill the missing data using the .isnull()
# We need the .loc for a location-based index, see pandas' documentation for more info
df.loc[(df.Age.isnull()),'AgeFill'] = median_age
df.head(8)
```

### 2 - Transform non-numeric data into numeric data

Let's map categorical data, such as male/female, to integers, such as 0/1.

```python
#Create a new column 'Gender' that maps the values of the column 'Sex' to nomerical values
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
```

```python
df.head(8)
```

### 3 - Drop columns we are not using

```python
#drop columns we wont use for now (and that are not convertible to numpy array)
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Age', 'PassengerId', 'Fare'], axis=1) 
```

```python
df.head()
```

### Now we can convert the pandas dataframe to numpy array, which will be the input file for our machine learning algorithm

```python
train_data = df.values         #convert to array
labels = train_data[:,0]       # First column is the labels (survived or not)
train_data = train_data[:, 1:] # All other columns are data to train the model
print train_data
```

### One last thing : we need to apply the exact same data treatment to the test data too!

```python
df = pd.read_csv('test.csv', header=0)
df['AgeFill'] = df['Age']
df.loc[(df.Age.isnull()),'AgeFill'] = median_age
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
passengerid = df.PassengerId.values #We also need to keep the passenger ids separately
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Age', 'PassengerId', 'Fare'], axis=1) 
test_data = df.values
```

# Machine learning : random forests

Python's most famous machine learning library is Sci-kit learn. Learn about it on their website : http://scikit-learn.org/stable/index.html

We will use the Random Forest Classifier, because it is usually robust to badly scaled data and to noise, so it's a good first choice for exploring. To learn more about random forests and decision trees, check out Wikipedia or this awesome post by Trevor Stephens : http://trevorstephens.com/post/72916401642/titanic-getting-started-with-r

```python
# Import the random forest package
from sklearn.ensemble import RandomForestClassifier 

# Create the random forest object which will include all the parameters
# for the fit
forest = RandomForestClassifier(n_estimators = 100)

# Fit the training data to the Survived labels and create the decision trees
forest = forest.fit(train_data,labels)

# Take the same decision trees and run it on the test data
output = forest.predict(test_data)
```

## Idea 3 : Train and test on the data we have and see if we gain accuracy by adding features

We need a metric for measuring the performance of the classifier

```python
def accuracy(labels, predictions):
    counts=0.
    for lab, pred in zip(labels, predictions):
        if lab==pred:
            counts+=1.
    return counts/len(labels)
```

The most convenient way to split our training data into smaller training and testing (cross-validation) sets is to use K-folds.

```python
from sklearn import cross_validation

cv = cross_validation.StratifiedKFold(labels,5)
results = []
for traincv, testcv in cv:
    forest.fit(train_data[traincv, :-1], labels[traincv]) #Not using special titles
    predictions = forest.predict(train_data[testcv, :-1])
    results.append(accuracy(labels[testcv], predictions))
print results
print np.array(results).mean()

# Get the importance, the higher the better
importance = forest.feature_importances_

# Print out the importance along with the name of the feature
for col, imp in zip(df.columns, importance):
    print col, imp
```
